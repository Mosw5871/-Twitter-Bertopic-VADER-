```python
import pandas as pd
NewtargetFile = "/Volumes/时伟SSD512G/扶贫数据-11列-英文推文筛选.xlsx"
```

```python
df = pd.read_excel(NewtargetFile) #读取excel文件
```

```python
df.head
```

```python
df_daterank = df.sort_values('date', ascending = True) #按日期排序
```

```python
df_daterank.head()
```

```python
#筛选出想要的日期范围2013-01-01到2021-12-31，赋给新的DataFrame：df_daterank_datefliter
df_daterank_datefliter = df_daterank[(df_daterank['date'] >= '2013-01-01') & 
                                        (df_daterank['date'] <= '2021-12-31')]
```

```python
df_daterank_datefliter = df_daterank_datefliter.drop(columns = ['Unnamed: 0'])
#reset_index()重置索引 https://www.jianshu.com/p/223de3248f90
#在获得新的index，原来的index变成数据列，保留在数据框中，不想保留原来的index的话可以使用参数drop=True，默认False。
df_daterank_datefliter = df_daterank_datefliter.reset_index(drop = True)
df_daterank_datefliter.head()
```
