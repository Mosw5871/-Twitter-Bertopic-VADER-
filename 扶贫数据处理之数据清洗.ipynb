```python
print('***************ÂºÄÂßãÂ§ÑÁêÜÊï∞ÊçÆÔºåÁõÆÂâçÊòØÁ¨¨‰∏ÄÊ≠•clean_specialtext(tweet)ÂáΩÊï∞ÁéØËäÇÔºåËØ∑ËÄêÂøÉÁ≠âÂæÖ***************')

#ÂØºÂÖ•nltkÁöÑÂàÜËØçÂ∑•ÂÖ∑word_tokenizeÂíånltkÁöÑÂÅúÁî®ËØçË°®stopwordsÔºåÈúÄË¶ÅËá™Ë°å‰∏ãËΩΩÂì¶
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

stopwords_list = stopwords.words('english')

#ÂÆö‰πâ‰∏Ä‰∏™ÂáΩÊï∞ÔºåÂ§ÑÁêÜTwitterÊñáÊú¨‰∏≠ÁöÑÁâπÊÆäÁ¨¶Âè∑Ôºö@xxx„ÄÅ&xxx„ÄÅÁΩëÂùÄ„ÄÅÂ§ö‰ΩôÁöÑÁ©∫Ê†ºÔºåÊúÄÂêéÂè™‰øùÁïôÊï∞Â≠óÂíåÂ≠óÊØç
def clean_specialtext(tweet):
    tweet = re.sub(r'@([^\s:]+)', ' ', tweet) #ÂéªÈô§@ÊèêÂèäÂÜÖÂÆπ
    tweet = re.sub(r'&([^\s:]+)', ' ', tweet) #ÂéªÈô§&Âèä‰πãÂêéÁöÑÂÜÖÂÆπ
    tweet = re.sub(r'[a-zA-z]+://[^\s]*', ' ', tweet) #ÂéªÈô§ÁΩëÂùÄ
    tweet = re.sub(r'[^0-9A-Za-z]', ' ', tweet) #‰ªÖ‰øùÁïôÊï∞Â≠óÂèäÂ≠óÊØç
    tweet = re.sub(r'(^\s*)|(\s*$)', '', tweet)  #ÂéªÈô§Â≠óÁ¨¶‰∏≤ÂâçÂêéÁ©∫Ê†º
    #tweet = re.sub(r'[^0-9A-Za-z]', ' ', tweet)
    return tweet

#Êó†Ê≥ïÁõ¥Êé•ÂØπÂéüÊù•ÁöÑdfËøõË°åÊìç‰ΩúÔºåÊâÄ‰ª•ËøôÈáåÈÄâÊã©ÂÖàÂ§çÂà∂‰∏Ä‰∏™resultÔºàdataframeÊ†ºÂºèÔºâÔºåÂú®Ëøô‰∏™Â§çÂà∂ÁâàÊú¨ÁöÑÊìç‰ΩúÂÆå‰πãÂêéÔºåÂêàÂπ∂ÂõûÂéüÊù•ÁöÑdf‰πã‰∏≠
result = df['tweet'].copy()
print('***************clean_specialtext(tweet)ÂáΩÊï∞Â∑≤ÁªèÂÆö‰πâÔºåÊ≠£Âú®ÊâßË°åËØ•ÂáΩÊï∞ÁöÑÂ§ÑÁêÜÔºåËØ∑ËÄêÂøÉÁ≠âÂæÖ***************')
#apply()ÂáΩÊï∞ÊòØÂØπdf‰∏≠ÁöÑÊØè‰∏ÄÂàóÁöÑÊØè‰∏Ä‰∏™Ê†ºÂ≠êÔºåÂ∫îÁî®‰∏Ä‰∏™ÂáΩÊï∞ÔºåËá™Âä®Âæ™ÁéØ
result = result.apply(lambda x: clean_specialtext(x))
#ÂêàÂπ∂ÂõûÂéüÊù•ÁöÑdf‰πã‰∏≠
df.insert(loc=4, column='tokenized_tweet', value = result)
```

```python
print('***************ÂºÄÂßãÂ§ÑÁêÜÊï∞ÊçÆÔºåÁõÆÂâçÊòØÁ¨¨‰∫åÊ≠•clean_stopwords_Lemmatization(tweet)ÂáΩÊï∞ÁéØËäÇÔºåËØ∑ËÄêÂøÉÁ≠âÂæÖ***************')

#ÂØºÂÖ•nltkÁöÑËØçÊÄßÊ†áÊ≥®Â∑•ÂÖ∑pos_tag„ÄÅËØçÊÄßÂ±ûÊÄßÂ∑•ÂÖ∑wordnetÂíåËØçÂΩ¢ËøòÂéüÂ∑•ÂÖ∑WordNetLemmatizer
from nltk.stem import WordNetLemmatizer
from nltk.tag import pos_tag
from nltk.corpus import wordnet
lemmatizer = WordNetLemmatizer()

#Ëøô‰∏ÄÊ≠•ÊòØ‰∏∫‰∫ÜÊõ¥Á≤æÁ°ÆËØçÂΩ¢ËøòÂéüÔºåÂØπËØçÊÄßËøõË°åÊ†áÊ≥®ÔºåÁ≤æÁ°ÆÁöÑËØçÊÄßÂØπËØçÂΩ¢ËøòÂéüÂæàÊúâÂøÖË¶Å 
#ÂèÇËÄÉhttps://stackoverflow.com/questions/61982023/using-wordnetlemmatizer-lemmatize-with-pos-tags-throws-keyerror
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return None

#ÂÆö‰πâÂéªÈô§ÂÅúÁî®ËØçÂíåËØçÂΩ¢ËøòÂéüÂáΩÊï∞ÔºåÊØè‰∏ÄÊ≠•ÈÉΩÊòØÊúâÈ°∫Â∫èÁöÑÔºå‰∏çËÉΩ‰π±
def clean_stopwords_Lemmatization(tweet):
    tweet = word_tokenize(tweet)
    #print(tweet)
    #print("*"*100)
    tweet = [word for word in tweet if word not in stopwords_list] #ÂéªÈô§ÂÅúÁî®ËØç
    tweet = pos_tag(tweet) #ËØçÊÄßÊ†áÊ≥®
    #ËØçÊÄßÊ†áÊ≥®Â•Ω‰πãÂêéÔºåÂàóË°®ÈáåÂèòÊàê‰∫ÜÔºàËØçÔºåËØçÊÄßÔºâÁöÑÂÖÉÁªÑÂΩ¢ÂºèÔºåword[0]ÊòØËØçÁöÑ‰ΩçÁΩÆÔºåÂØπword[0]ËøõË°åËØçÂΩ¢ËøòÂéüÔºå‰æùÊçÆÊòØ‰∏äÈù¢ÁöÑÂáΩÊï∞Âíåword[1]
    tweet = [lemmatizer.lemmatize(word[0], pos = get_wordnet_pos(word[1])) if get_wordnet_pos(word[1]) else word[0] for word in tweet]
    tweet = [word for word in tweet if len(word) > 2] #‰∏¢ÂºÉÈïøÂ∫¶Â∞è‰∫éÁ≠â‰∫é2ÁöÑÂ≠óÁ¨¶‰∏≤
    tweet = ' '.join(tweet) #Â∞ÜtweetÂàóË°®‰∏≠ÁöÑÂ≠óÁ¨¶‰∏≤ÈáçÊñ∞ÊãºÊé•Ëµ∑Êù•ÔºåÁî®Á©∫Ê†ºÂàÜÈöî
    #print(tweet)
    #print("#"*100)
    return tweet

#Êó†Ê≥ïÁõ¥Êé•ÂØπÂéüÊù•ÁöÑdfËøõË°åÊìç‰ΩúÔºåÊâÄ‰ª•ËøôÈáåÈÄâÊã©ÂÖàÂ§çÂà∂‰∏Ä‰∏™result2ÔºàdataframeÊ†ºÂºèÔºâÔºåÂú®Ëøô‰∏™Â§çÂà∂ÁâàÊú¨ÁöÑÊìç‰ΩúÂÆå‰πãÂêéÔºåÂêàÂπ∂ÂõûÂéüÊù•ÁöÑdf‰πã‰∏≠
result2 = df['tokenized_tweet'].copy()
print('***************clean_stopwords_Lemmatization(tweet)ÂáΩÊï∞Â∑≤ÁªèÂÆö‰πâÔºåÊ≠£Âú®ÊâßË°åËØ•ÂáΩÊï∞ÁöÑÂ§ÑÁêÜÔºåËØ∑ËÄêÂøÉÁ≠âÂæÖ***************')
#apply()ÂáΩÊï∞ÊòØÂØπdf‰∏≠ÁöÑÊØè‰∏ÄÂàóÁöÑÊØè‰∏Ä‰∏™Ê†ºÂ≠êÔºåÂ∫îÁî®‰∏Ä‰∏™ÂáΩÊï∞ÔºåËá™Âä®Âæ™ÁéØ
result2 = result2.apply(lambda x: clean_stopwords_Lemmatization(x))
#ÂêàÂπ∂ÂõûÂéüÊù•ÁöÑdf‰πã‰∏≠
df.insert(loc=4, column='clean_tweet', value = result2)
```
