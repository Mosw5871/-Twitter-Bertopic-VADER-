```python
df = df.drop(columns = ['Unnamed: 0'])
print('**********åˆ é™¤ä¸å¿…è¦çš„åˆ—**********')
df
```

```python
df['tweet'] = df['tweet'].str.lower() #ç»Ÿä¸€ä¸ºå°å†™å­—æ¯
print('**********å…¨éƒ¨æ–‡æœ¬å·²ç»è½¬æ¢æˆå°å†™æ ¼å¼**********')
df
```

```python
print('***************å¼€å§‹å¤„ç†æ•°æ®ï¼Œç›®å‰æ˜¯ç¬¬ä¸€æ­¥clean_specialtext(tweet)å‡½æ•°ç¯èŠ‚ï¼Œè¯·è€å¿ƒç­‰å¾…***************')

#å¯¼å…¥nltkçš„åˆ†è¯å·¥å…·word_tokenizeå’Œnltkçš„åœç”¨è¯è¡¨stopwordsï¼Œéœ€è¦è‡ªè¡Œä¸‹è½½å“¦
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

stopwords_list = stopwords.words('english')

#å®šä¹‰ä¸€ä¸ªå‡½æ•°ï¼Œå¤„ç†Twitteræ–‡æœ¬ä¸­çš„ç‰¹æ®Šç¬¦å·ï¼š@xxxã€&xxxã€ç½‘å€ã€å¤šä½™çš„ç©ºæ ¼ï¼Œæœ€ååªä¿ç•™æ•°å­—å’Œå­—æ¯
def clean_specialtext(tweet):
    tweet = re.sub(r'@([^\s:]+)', ' ', tweet) #å»é™¤@æåŠå†…å®¹
    tweet = re.sub(r'&([^\s:]+)', ' ', tweet) #å»é™¤&åŠä¹‹åçš„å†…å®¹
    tweet = re.sub(r'[a-zA-z]+://[^\s]*', ' ', tweet) #å»é™¤ç½‘å€
    tweet = re.sub(r'[^0-9A-Za-z]', ' ', tweet) #ä»…ä¿ç•™æ•°å­—åŠå­—æ¯
    tweet = re.sub(r'(^\s*)|(\s*$)', '', tweet)  #å»é™¤å­—ç¬¦ä¸²å‰åç©ºæ ¼
    #tweet = re.sub(r'[^0-9A-Za-z]', ' ', tweet)
    return tweet

#æ— æ³•ç›´æ¥å¯¹åŸæ¥çš„dfè¿›è¡Œæ“ä½œï¼Œæ‰€ä»¥è¿™é‡Œé€‰æ‹©å…ˆå¤åˆ¶ä¸€ä¸ªresultï¼ˆdataframeæ ¼å¼ï¼‰ï¼Œåœ¨è¿™ä¸ªå¤åˆ¶ç‰ˆæœ¬çš„æ“ä½œå®Œä¹‹åï¼Œåˆå¹¶å›åŸæ¥çš„dfä¹‹ä¸­
result = df['tweet'].copy()
print('***************clean_specialtext(tweet)å‡½æ•°å·²ç»å®šä¹‰ï¼Œæ­£åœ¨æ‰§è¡Œè¯¥å‡½æ•°çš„å¤„ç†ï¼Œè¯·è€å¿ƒç­‰å¾…***************')
#apply()å‡½æ•°æ˜¯å¯¹dfä¸­çš„æ¯ä¸€åˆ—çš„æ¯ä¸€ä¸ªæ ¼å­ï¼Œåº”ç”¨ä¸€ä¸ªå‡½æ•°ï¼Œè‡ªåŠ¨å¾ªç¯
result = result.apply(lambda x: clean_specialtext(x))
#åˆå¹¶å›åŸæ¥çš„dfä¹‹ä¸­
df.insert(loc=4, column='tokenized_tweet', value = result)
```

```python
print('***************å¼€å§‹å¤„ç†æ•°æ®ï¼Œç›®å‰æ˜¯ç¬¬äºŒæ­¥clean_stopwords_Lemmatization(tweet)å‡½æ•°ç¯èŠ‚ï¼Œè¯·è€å¿ƒç­‰å¾…***************')

#å¯¼å…¥nltkçš„è¯æ€§æ ‡æ³¨å·¥å…·pos_tagã€è¯æ€§å±æ€§å·¥å…·wordnetå’Œè¯å½¢è¿˜åŸå·¥å…·WordNetLemmatizer
from nltk.stem import WordNetLemmatizer
from nltk.tag import pos_tag
from nltk.corpus import wordnet
lemmatizer = WordNetLemmatizer()

#è¿™ä¸€æ­¥æ˜¯ä¸ºäº†æ›´ç²¾ç¡®è¯å½¢è¿˜åŸï¼Œå¯¹è¯æ€§è¿›è¡Œæ ‡æ³¨ï¼Œç²¾ç¡®çš„è¯æ€§å¯¹è¯å½¢è¿˜åŸå¾ˆæœ‰å¿…è¦ 
#å‚è€ƒhttps://stackoverflow.com/questions/61982023/using-wordnetlemmatizer-lemmatize-with-pos-tags-throws-keyerror
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return None

#å®šä¹‰å»é™¤åœç”¨è¯å’Œè¯å½¢è¿˜åŸå‡½æ•°ï¼Œæ¯ä¸€æ­¥éƒ½æ˜¯æœ‰é¡ºåºçš„ï¼Œä¸èƒ½ä¹±
def clean_stopwords_Lemmatization(tweet):
    tweet = word_tokenize(tweet)
    #print(tweet)
    #print("*"*100)
    tweet = [word for word in tweet if word not in stopwords_list] #å»é™¤åœç”¨è¯
    tweet = pos_tag(tweet) #è¯æ€§æ ‡æ³¨
    #è¯æ€§æ ‡æ³¨å¥½ä¹‹åï¼Œåˆ—è¡¨é‡Œå˜æˆäº†ï¼ˆè¯ï¼Œè¯æ€§ï¼‰çš„å…ƒç»„å½¢å¼ï¼Œword[0]æ˜¯è¯çš„ä½ç½®ï¼Œå¯¹word[0]è¿›è¡Œè¯å½¢è¿˜åŸï¼Œä¾æ®æ˜¯ä¸Šé¢çš„å‡½æ•°å’Œword[1]
    tweet = [lemmatizer.lemmatize(word[0], pos = get_wordnet_pos(word[1])) if get_wordnet_pos(word[1]) else word[0] for word in tweet]
    tweet = [word for word in tweet if len(word) > 2] #ä¸¢å¼ƒé•¿åº¦å°äºç­‰äº2çš„å­—ç¬¦ä¸²
    tweet = ' '.join(tweet) #å°†tweetåˆ—è¡¨ä¸­çš„å­—ç¬¦ä¸²é‡æ–°æ‹¼æ¥èµ·æ¥ï¼Œç”¨ç©ºæ ¼åˆ†éš”
    #print(tweet)
    #print("#"*100)
    return tweet

#æ— æ³•ç›´æ¥å¯¹åŸæ¥çš„dfè¿›è¡Œæ“ä½œï¼Œæ‰€ä»¥è¿™é‡Œé€‰æ‹©å…ˆå¤åˆ¶ä¸€ä¸ªresult2ï¼ˆdataframeæ ¼å¼ï¼‰ï¼Œåœ¨è¿™ä¸ªå¤åˆ¶ç‰ˆæœ¬çš„æ“ä½œå®Œä¹‹åï¼Œåˆå¹¶å›åŸæ¥çš„dfä¹‹ä¸­
result2 = df['tokenized_tweet'].copy()
print('***************clean_stopwords_Lemmatization(tweet)å‡½æ•°å·²ç»å®šä¹‰ï¼Œæ­£åœ¨æ‰§è¡Œè¯¥å‡½æ•°çš„å¤„ç†ï¼Œè¯·è€å¿ƒç­‰å¾…***************')
#apply()å‡½æ•°æ˜¯å¯¹dfä¸­çš„æ¯ä¸€åˆ—çš„æ¯ä¸€ä¸ªæ ¼å­ï¼Œåº”ç”¨ä¸€ä¸ªå‡½æ•°ï¼Œè‡ªåŠ¨å¾ªç¯
result2 = result2.apply(lambda x: clean_stopwords_Lemmatization(x))
#åˆå¹¶å›åŸæ¥çš„dfä¹‹ä¸­
df.insert(loc=4, column='clean_tweet', value = result2)
```
